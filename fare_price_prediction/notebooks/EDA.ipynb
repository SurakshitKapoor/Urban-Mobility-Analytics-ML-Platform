```python
# --- Standardize city_id (strip spaces, uppercase) ---
tables = [
    monthly_target_trips,
    monthly_target_new_passengers,
    fact_trips,
    fact_passenger_summary,
    dim_repeat_trip_distribution,
    dim_city,
    city_target_passenger_rating
]

for t in tables:
    if 'city_id' in t.columns:
        t['city_id'] = t['city_id'].str.strip().str.upper()

# --- Standardize month/date formats ---
monthly_target_trips['month'] = pd.to_datetime(monthly_target_trips['month'])
monthly_target_new_passengers['month'] = pd.to_datetime(monthly_target_new_passengers['month'])
fact_passenger_summary['month'] = pd.to_datetime(fact_passenger_summary['month'])
dim_repeat_trip_distribution['month'] = pd.to_datetime(dim_repeat_trip_distribution['month'])

fact_trips['date'] = pd.to_datetime(fact_trips['date'])
dim_date['date'] = pd.to_datetime(dim_date['date'])
dim_date['start_of_month'] = pd.to_datetime(dim_date['start_of_month'])
```

Passenger features

```python
# base passenger table

passenger_features = fact_passenger_summary.copy()

# create repeat ratio
passenger_features['repeat_ratio'] = (
    passenger_features['repeat_passengers'] / passenger_features['total_passengers']
)
```

```python
# repeat trip distribution table

repeat_pivot = dim_repeat_trip_distribution.pivot_table(
    index=['month', 'city_id'],
    columns='trip_count',
    values='repeat_passenger_count',
    aggfunc='sum'
).reset_index()

repeat_pivot = repeat_pivot.fillna(0)

# prefix columns
repeat_pivot = repeat_pivot.add_prefix("rt_")
repeat_pivot = repeat_pivot.rename(columns={"rt_month": "month", "rt_city_id": "city_id"})
```

```python
# merge alltable

passenger_features = passenger_features.merge(repeat_pivot, on=['month', 'city_id'], how='left')
```

Aggergations from Fact 

```python
trip_features = (
    fact_trips
    .assign(month=lambda x: x['date'].values.astype('datetime64[M]'))
    .groupby(['city_id', 'month'])
    .agg(
        total_trips=('trip_id', 'count'),
        avg_fare=('fare_amount', 'mean'),
        avg_distance=('distance_travelled(km)', 'mean'),
        avg_passenger_rating=('passenger_rating', 'mean'),
        avg_driver_rating=('driver_rating', 'mean')
    )
    .reset_index()
)
```

City Features

```python
city_features = dim_city.merge(
    city_target_passenger_rating,
    on='city_id',
    how='left'
)
```

target features

```python
target_features = monthly_target_trips.merge(
    monthly_target_new_passengers,
    on=['city_id', 'month'],
    how='outer'
)
```

merge evrything together: 

Golden order:

passenger_features

trip_features

city_features

target_features

```python
final = passenger_features.merge(
    trip_features, on=['city_id','month'], how='left'
).merge(
    city_features, on='city_id', how='left'
).merge(
    target_features, on=['city_id','month'], how='left'
)
```

compute target gaps

```python
final['passenger_target_gap'] = (
    final['total_passengers'] - final['target_new_passengers']
)

final['trip_target_gap'] = (
    final['total_trips'] - final['total_target_trips']
)
```

final cleaning

```python
final = final.sort_values(['city_id', 'month']).reset_index(drop=True)
```

add new_passenger_ratio

```python
final["new_passenger_ratio"] = final["new_passengers"] / final["total_passengers"]
```
